```python
df = pd.read_excel("../data/Basic research participation information .xlsx")
```

```python
df.dropna(thresh=df.shape[1]*0.6,axis=0,inplace=True)
```

```python
from datetime import date,datetime

def calculate_age(birthdate):
    bdate = datetime.strptime(str(birthdate), "%Y%m%d").date()
    today = date.today()
    age = today.year - bdate.year - ((today.month, today.day) < (bdate.month, bdate.day))
    return age

df["Age"]=df["Date of birth"].apply(calculate_age)
```
